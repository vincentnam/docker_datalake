### Importation des bibliothèques nécessaires

```python
import cv2 
import os
import time
import numpy as np
import pandas as pd
```

### Instanciation d'objet descriptors

```python
import descriptors
d = descriptors.Descriptors(nbr_class = 5 , nbr_images =500)
```

### Récupération des paths

```python
(Tempo,_) = os.path.split(os.getcwd())
PATH_PROJECT,DEV = os.path.split(Tempo)
PATH_DEV = os.path.join(PATH_PROJECT,'Dev')
PATH_DATA = os.path.join(PATH_PROJECT,'Data')
PATH_DATA_IMAGES = os.path.join(PATH_DATA,'Data Images')
DATA_CBIRH_5_500 = os.path.join(PATH_DATA_IMAGES,'DATA_CBIRH_5_500')
PATH_DATA_CSV = os.path.join(os.path.join(os.path.join(PATH_DATA,'Data Descriptors'),'Data Feautures_5_500'),'Three descriptors')
LIST_CLASS = os.listdir(DATA_CBIRH_5_500)

PATH_LIST =[]
for f in LIST_CLASS :
    PATH_LIST.append(os.path.join(DATA_CBIRH_5_500,f))
```

## Create data features GCM_DCD_HM



```python
columns = ['name_image','dissimilarity','correlation', 'contrast','homogeneity','energy',
           'l_1','u_1','v_1','percent_1','l_2','u_2','v_2','percent_2','l_3','u_3','v_3','percent_3',
           'm_1', 'm_2', 'm_3', 'm_4', 'm_5', 'm_6', 'm_7','label']
data = d.create_data_feautures_3_descriptors(PATH_LIST,columns,PATH_DATA_CSV, 'DF_DCD_GCM_HM.csv',d.DOMINANT_COLOR_DESCRIPTOR, d.DESCRIPTOR_TEXTURE_MATRIX_COCCURRENCE, d.descriprot_moments_Hu)
data.head(10)
```

## Create data features CH_TH_HM

```python
columns = ['name_img','mean_color_1', 'var_color_1', 'kurtosis_color_1', 'entropy_color_1',
                        'mean_color_2', 'var_color_2', 'kurtosis_color_2', 'entropy_color_2',
                        'mean_color_3', 'var_color_3', 'kurtosis_color_3', 'entropy_color_3',
                     'mean_color_gray', 'var_color_gray', 'kurtosis_color_gray', 'entropy_color_gray',
                       'm_1', 'm_2', 'm_3', 'm_4', 'm_5', 'm_6', 'm_7','label']
data = d.create_data_feautures_3_descriptors(PATH_LIST,columns,PATH_DATA_CSV, 'DF_CH_TH_HM.csv',d.descriprot_color_hitogramme, d.descriprot_texture_hitogramme, d.descriprot_moments_Hu)
data.head(10)
```
